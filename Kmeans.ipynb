make necessary imports

```python
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
import numpy as np
```

load dataset

```python
f = open("bc.train.0.txt")
data = np.loadtxt(f)
```

```python
X = data[:,1:]
Y = data[:,0]
```

perform clustering

```python
clustering = KMeans(n_clusters=2,init='random').fit(X)
```

find error

```python
err = 0 
for i in range(len(X)):
    if (clustering.labels_[i] != Y[i]):
        err+=1
err/= len(X)
print(err)
```

```python
for i in range(len(X)):
    print(clustering.labels_[i],Y[i])
```
